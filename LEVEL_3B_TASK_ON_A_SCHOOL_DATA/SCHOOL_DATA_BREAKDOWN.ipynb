 ## <span style="color:pink;"><u><h3>STEP 5: GROUPING AND AGGREGATION.</u></h3></span>

 ## <span style="color:ORANGE;">Use df.groupby() to find the average age of students in each grade.</span>

```python
df_avg_age = df.groupby("Grade")["Age"].mean()
df_avg_age
#df.groupby("Grade")  Groups the dataset by the "Grade" column. AND .mean()  Calculates the average (mean) age for each grade.
```

```python
df_avg_age = df.groupby("Grade", as_index=False)["Age"].mean()
df_avg_age
#as_index=False ensures "Grade" remains a column instead of becoming an index.
```

 ## <span style="color:orange;">Use df.groupby() to find the maximum marks in Math for each grade.</span>

```python
Y=df.groupby("Grade")["Math"].max()
Y
#Here it looks at  person who has the HIGHEST score in MATH but has a GRADE A then moves to B,C and D
```

 ## <span style="color:orange;">Use df.groupby() to find the minimum marks in ENGLISH for each grade.</span>

```python
Z=df.groupby("Grade")["English"].min()
Z
#Here it looks at  person who has the LOWEST score in MATH but has a GRADE A then moves to B,C and D
```
