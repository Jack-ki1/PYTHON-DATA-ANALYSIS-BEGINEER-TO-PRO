### **1. Matplotlib: The Foundational Library**
Matplotlib is designed for creating publication-quality 2D and 3D charts, aiming to replicate a **Matlab-like environment**.

*   **Core Architecture**: It consists of three layers: the **Backend layer** (low-level implementation), the **Artist layer** (hierarchy of all chart components like Figure, Axes, and Axis), and the **Scripting layer** (the `pyplot` module).
*   **Key Plotting Commands**:
    *   `plt.plot()`: The default function for **line charts**.
    *   `plt.scatter()`: Creates **scatter plots** for visualizing relationships between continuous variables.
    *   `plt.bar()` / `plt.barh()`: Generates **vertical and horizontal bar charts**.
    *   `plt.hist()`: Plots **histograms** to show frequency distributions; it also returns a tuple of calculated bin values.
    *   `plt.pie()`: Used for **pie charts** to compare parts of a whole.
    *   `plt.stackplot()`: Visualizes the **frequency of activities** stacked over a time interval.
    *   `plt.contour()` / `plt.contourf()`: Renders **contour maps** for 3D surfaces.
    *   `plt.boxplot()`: Summarizes data using **medians, quartiles, and outliers**.
*   **Customization and Decoration**:
    *   **Text and Titles**: `plt.title()`, `plt.xlabel()`, `plt.ylabel()`, and `plt.suptitle()` for figure-level titles.
    *   **Ticks and Ticks Labels**: `plt.xticks()` and `plt.yticks()` allow manual placement of ticks and support **LaTeX mathematical expressions** when enclosed in `$` symbols.
    *   **Legends and Grids**: `plt.legend()` identifies data series, and `plt.grid(True)` improves readability by adding a background grid.
    *   **Annotations**: `plt.text()` places text at specific coordinates, and `plt.annotate()` can add **labels with arrows** to highlight data points.
*   **3D Visualization (`mplot3d`)**:
    *   Requires initializing an `Axes3D` object.
    *   `ax.plot_surface()`: Renders **3D surfaces**.
    *   `ax.scatter()`: Creates **3D scatter plots** to identify data clusters.
    *   `ax.bar()`: Produces **3D bar charts**.
*   **Figure Management**:
    *   `plt.subplot()` / `plt.subplots()`: Divides the drawing area into a **grid of subplots**.
    *   `plt.savefig()`: Saves figures in formats like **PNG, SVG, or PDF**.

### **2. Seaborn: High-Level Statistical Graphics**
Seaborn is a declarative library built on top of Matplotlib that provides a higher-level interface to create complex statistical visualizations with **fewer lines of code**.

*   **Univariate Distribution Plots**:
    *   `sns.histplot()`: Creates histograms, often combined with a **Kernel Density Estimate (KDE)**.
    *   `sns.kdeplot()`: Generates a **continuous density plot**.
    *   `sns.rugplot()`: Adds marginal tick marks to represent individual data points.
    *   `sns.displot()`: A flexible function for various distributions that returns a **FacetGrid**.
*   **Categorical and Bivariate Plots**:
    *   `sns.barplot()`: Plots means with **automatic 95% confidence interval error bars**.
    *   `sns.countplot()`: A bar plot specifically for counting occurrences of categorical variables.
    *   `sns.boxplot()` / `sns.violinplot()`: Visualizes statistical distributions; violin plots include a **KDE** within the box plot structure.
    *   `sns.stripplot()` / `sns.swarmplot()`: Scatter plots for categorical data; `swarmplot` **prevents point overlap**.
    *   `sns.regplot()` / `sns.lmplot()`: Combines scatter plots with a **fitted linear regression line**.
*   **Matrix and Multivariate Plots**:
    *   `sns.jointplot()`: Combines bivariate scatter/hexbin plots with **univariate marginal distributions**.
    *   `sns.pairplot()` / `sns.PairGrid()`: Visualizes **all pairwise relationships** in a dataset simultaneously.
    *   `sns.heatmap()`: Uses color to visualize **patterns or correlation matrices**.
    *   `sns.catplot()` / `sns.relplot()`: Unified interfaces for creating **facet grids** to compare categorical subsets.
*   **Aesthetics and Styles**:
    *   `sns.set_style()`: Changes background themes (e.g., **'darkgrid', 'whitegrid', 'ticks'**).
    *   `sns.set_context()`: Scales plot elements for different uses (**'paper', 'notebook', 'talk', 'poster'**).
    *   `sns.set_palette()`: Customizes **color schemes**, such as the colorblind-friendly 'viridis'.

### **3. Integration with Pandas**
Pandas objects include internal plotting methods that act as **wrappers around Matplotlib**. Commands like `df.plot.bar()`, `df.plot.hist()`, `df.plot.box()`, and `df.plot.scatter()` allow users to generate standard Matplotlib charts directly from DataFrames while automatically utilizing column and index labels.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

## <span style="color:orange;">Now here is an indepth task based breakdown of seaborn and matplotlib</span>

```python
df=pd.read_csv("SOO3_DATA.csv")
```

### **the process below is to make ploting timelines possible hence the creation of columns for yearmonth and months specifically**

```python
# Ensure Order_Date is datetime format
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
# Extract Year-Month from Order_Date
df['YearMonth'] = df['Order_Date'].dt.to_period('M').astype(str)
# Extract month
df['Month'] = df['Order_Date'].dt.month
```

### Basic Visualizations

```python
#Plot a histogram of the Sales_Amount column.
plt.figure(figsize=(10, 6))
plt.hist(df["Sales_Amount"], bins=8, color="skyblue", edgecolor="C1")#hist is for histogram AND bins controls number of boxes you see
plt.title("histogram of the Sales_Amount")
plt.xlabel('SALES DISTRIBUTION')
plt.ylabel('SALES')
plt.show()
```

```python
#Create a bar chart showing the total sales for each Product_Category.
plt.figure(figsize=(12, 8))
df.groupby("Product_Category")["Sales_Amount"].sum().plot(kind='bar', color='C2')
plt.title("Total sales for each Product_Category.")
plt.xlabel('Product_Category')
plt.ylabel('Sales_Amount')
plt.show()
```

```python
#Visualize the distribution of Profit using a box plot.
plt.figure(figsize=(10,6))
sns.boxplot(df["Profit"], color='pink')
plt.title("Distribution of Profit")
plt.show()
```

```python
#Plot a pie chart showing the proportion of sales by Region.
plt.figure(figsize=(12, 8))
df.groupby("Region")["Sales_Amount"].sum().plot(kind='pie')
plt.title("The proportion of sales by Region")
plt.xlabel('Region')
plt.ylabel('Sales_Amount')
plt.show()
```

```python
#Create a line plot of total sales over time (using Order_Date).
plt.figure(figsize=(12, 8))
df.groupby("YearMonth")["Sales_Amount"].sum().plot(kind='line', color='orange')
plt.title("total sales over time (using Order_Date)")
plt.xlabel('Order_Date')
plt.ylabel('Sales_Amount')
plt.show()
```

```python
#Plot a scatter plot between Sales_Amount and Profit.
plt.figure(figsize=(12, 8))
plt.scatter(x=df["Sales_Amount"], y=df["Profit"], color='orange')
plt.title('Sales_Amount')
plt.xlabel('Sales_Amount')
plt.ylabel('Profit')
plt.show()
```

```python
#Visualize the distribution of Quantity using a histogram.
plt.figure(figsize=(12, 8))
plt.hist(df["Quantity"], bins=10, color="skyblue", edgecolor="grey")#hist is for histogram AND bins controls number of boxes you see
plt.title("Distribution of Quantity")
#plt.xlabel()
#plt.ylabel()
plt.show()
```

```python
#Visualize the correlation between numeric columns using a heatmap.
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='tab20', fmt='.2f')
plt.title('Correlation Heatmap of numeric columns')
plt.show() 
```

```python
#Visualize the monthly sales for each Product_Category using subplots
# Group by Month and Category, then sum Quantity (or use Revenue if available)
monthly_sales = df.groupby(['YearMonth', 'Product_Category'])['Quantity'].sum().reset_index()
# Plot with FacetGrid
g = sns.FacetGrid(monthly_sales, col="Product_Category", col_wrap=2, height=4, sharex=True, sharey=False)
g.map_dataframe(sns.lineplot, x="YearMonth", y="Quantity", marker="o")
g.set_titles("{col_name} Sales")
g.set_xticklabels(rotation=45)
plt.tight_layout()
plt.show()
```

```python
##Create a heatmap showing sales by month and Region.
# Create a pivot table: Rows = Month, Columns = Region, Values = Sales_Amount
heat_data = df.pivot_table(values='Sales_Amount', index='Month', columns='Region', aggfunc='sum')

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heat_data, annot=True, cmap='tab20', fmt='.2f')
plt.title('Monthly Sales by Region')
plt.show()
```

```python
#Plot the moving average of Sales_Amount over time. 
# Sort by date
df = df.sort_values('Order_Date')

# Create a 7-day moving average column
df['Moving_Avg'] = df['Sales_Amount'].rolling(window=7).mean()

# Plot using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Order_Date', y='Sales_Amount', label='Daily Sales')        # Original sales
sns.lineplot(data=df, x='Order_Date', y='Moving_Avg', label='7-Day Moving Avg')     # Moving average
plt.title('Sales Amount with 7-Day Moving Average')
plt.xlabel('Order_Date')
plt.ylabel('Sales Amount')
plt.legend()
plt.tight_layout()
plt.show()
#rolling(window=7) computes the moving average over the last 7 rows (days).You can change window=7 to another number like 30 for monthly trends.
```
