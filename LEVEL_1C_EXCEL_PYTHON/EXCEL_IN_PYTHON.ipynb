### **4. Cell-Level Control with `openpyxl`**

#Use `openpyxl` when you need **precise formatting**, **formulas**, or **cell-by-cell editing**.

#### **Read Cell Values & Metadata**

```python
from openpyxl import load_workbook

wb = load_workbook("data.xlsx")
ws = wb.active  # or wb["Sheet1"]

# Read cell value
value = ws["A1"].value
# Or by row/col
value = ws.cell(row=1, column=1).value

# Get cell style info
font = ws["A1"].font
fill = ws["A1"].fill
```

#### **Write & Format Cells**

```python
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side

wb = Workbook()
ws = wb.active

# Write values
ws["A1"] = "Hello"
ws.cell(row=2, column=1, value=42)

# Apply formatting
ws["A1"].font = Font(bold=True, color="FF0000")
ws["A1"].fill = PatternFill(start_color="FFFF00", fill_type="solid")
ws["A1"].alignment = Alignment(horizontal="center")

# Add border
thin_border = Border(
    left=Side(style='thin'),
    right=Side(style='thin'),
    top=Side(style='thin'),
    bottom=Side(style='thin')
)
ws["A1"].border = thin_border

wb.save("formatted.xlsx")
```

#### **Insert Formulas**

```python
ws["C1"] = "=A1+B1"
ws["D1"] = "=SUM(A1:A10)"
```

#### **Adjust Column Width & Row Height**

```python
ws.column_dimensions["A"].width = 20
ws.row_dimensions[1].height = 30
```

#### **Merge Cells**

```python
from openpyxl.utils import get_column_letter

ws.merge_cells("A1:C1")  # Merge A1 to C1
# Or dynamically:
ws.merge_cells(start_row=1, start_column=1, end_row=1, end_column=3)
```

### **5. Advanced Formatting with `openpyxl`**

#### **Conditional Formatting**

```python
from openpyxl.formatting.rule import ColorScaleRule

# Color scale: red-yellow-green
ws.conditional_formatting.add(
    "B2:B100",
    ColorScaleRule(
        start_type='min', start_color='FF0000',
        mid_type='percentile', mid_value=50, mid_color='FFFF00',
        end_type='max', end_color='00FF00'
    )
)
```

#### **Number Formatting**

```python
ws["B2"].number_format = "#,##0.00"      # 1,234.56
ws["C2"].number_format = "0.00%"         # 50.00%
ws["D2"].number_format = "yyyy-mm-dd"    # 2025-01-27
```

#### **Add Comments**

```python
from openpyxl.comments import Comment

ws["A1"].comment = Comment("This is important!", "Author")
```

### **6. Creating Charts with `openpyxl`**

```python
from openpyxl.chart import BarChart, Reference

# Assume data in A1:B4
data = Reference(ws, min_col=2, min_row=1, max_col=2, max_row=4)
categories = Reference(ws, min_col=1, min_row=2, max_row=4)

chart = BarChart()
chart.add_data(data, titles_from_data=True)
chart.set_categories(categories)
chart.title = "Sales by Region"

ws.add_chart(chart, "E2")  # Place chart at E2
```

> ðŸ”” **Note**: `openpyxl` supports basic charts. For advanced visuals, use `pandas` + `matplotlib`/`plotly` and embed as images.

### **7. High-Performance Writing with `xlsxwriter`**

Use `xlsxwriter` when you need **speed**, **large files**, or **advanced Excel features** (e.g., sparklines, data validation).

#### **Basic Write**

```python
import xlsxwriter

workbook = xlsxwriter.Workbook("fast_output.xlsx")
worksheet = workbook.add_worksheet()

# Write data
worksheet.write("A1", "Name")
worksheet.write("B1", "Score")
worksheet.write_row("A2", ["Alice", 95])
worksheet.write_column("A3", ["Bob", "Charlie"])

workbook.close()
```

### **Formatting with `xlsxwriter`**

#### Define format

```python
bold = workbook.add_format({"bold": True, "font_color": "red"})
currency = workbook.add_format({"num_format": "$#,##0.00"})

# Apply
worksheet.write("A1", "Total", bold)
worksheet.write("B2", 1234.5, currency)
```

#### **Data Validation (Dropdowns)**

```python
worksheet.data_validation(
    "C2:C100",
    {"validate": "list", "source": ["Yes", "No", "Maybe"]}
)
```

#### **Sparklines**

```python
worksheet.add_sparkline(
    "E2",
    {"range": "B2:D2", "type": "column"}
)
```

### **8. Real-World Examples**

#### **Example 1: Clean & Export Report**

```python
# Read messy data
df = pd.read_excel("raw_sales.xlsx", skiprows=3)

# Clean
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
df = df.dropna(subset=["order_id"])
df["order_date"] = pd.to_datetime(df["order_date"])

# Export with formatting
with pd.ExcelWriter("clean_report.xlsx", engine="xlsxwriter") as writer:
    df.to_excel(writer, sheet_name="Clean Data", index=False)
    
    # Get workbook/worksheet objects
    workbook = writer.book
    worksheet = writer.sheets["Clean Data"]
    
    # Add auto-filter
    worksheet.autofilter(0, 0, len(df), len(df.columns)-1)
    
    # Set column width
    for i, col in enumerate(df.columns):
        width = max(df[col].astype(str).map(len).max(), len(col)) + 2
        worksheet.set_column(i, i, width)
```

### **9. Performance Tips**

| Task | Recommendation |
|------|----------------|
| **Reading large files** | Use `read_excel(..., usecols=..., nrows=...)` to limit data |
| **Writing large DataFrames** | Use `xlsxwriter` (faster than `openpyxl` for write-only) |
| **Repeated operations** | Avoid cell-by-cell writes; use `write_row()` or `DataFrame.to_excel()` |
| **Memory usage** | For huge files, consider `pylightxl` (read-only, low memory) |

---

### **10. Common Pitfalls & Solutions**

| Problem | Solution |
|--------|--------|
| **Dates read as integers** | Use `parse_dates` in `pd.read_excel()` or convert with `pd.to_datetime()` |
| **Formulas not calculated** | Excel must recalculate; `openpyxl` reads last computed value |
| **Unicode errors** | Ensure file encoding is UTF-8; use `engine="openpyxl"` |
| **File locked for writing** | Close file in Excel before running script |
| **Slow performance** | Avoid `iter_rows()` for large datasets; use pandas instead |

---

### **11. When to Use Which Library?**

- **Quick data analysis?** â†’ `pandas.read_excel()`
- **Need to preserve formatting?** â†’ `openpyxl`
- **Generating reports from scratch?** â†’ `xlsxwriter`
- **Working with old `.xls` files?** â†’ `xlrd` (read-only) or `pyxlsb` for `.xlsb`

### **Final Project: Automated Sales Dashboard Generator**

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill
import datetime

# 1. Load data
sales = pd.read_excel("sales_raw.xlsx")
sales["Date"] = pd.to_datetime(sales["Date"])

# 2. Summarize
summary = sales.groupby("Region").agg(
    Total_Sales=("Amount", "sum"),
    Avg_Order=("Amount", "mean"),
    Orders=("OrderID", "count")
).round(2)

# 3. Write to Excel with formatting
output_file = f"Sales_Report_{datetime.date.today()}.xlsx"
summary.to_excel(output_file, sheet_name="Summary")

# 4. Enhance with openpyxl
wb = load_workbook(output_file)
ws = wb["Summary"]

# Header style
header_fill = PatternFill(start_color="4F81BD", end_color="4F81BD", fill_type="solid")
header_font = Font(color="FFFFFF", bold=True)

for col in ws.iter_cols(min_row=1, max_row=1):
    for cell in col:
        cell.fill = header_fill
        cell.font = header_font

# Auto-fit columns
for col in ws.columns:
    max_length = 0
    column = col[0].column_letter
    for cell in col:
        try:
            if len(str(cell.value)) > max_length:
                max_length = len(str(cell.value))
        except:
            pass
    adjusted_width = min(max_length + 2, 50)
    ws.column_dimensions[column].width = adjusted_width

wb.save(output_file)
print(f"âœ… Report saved: {output_file}")
```
