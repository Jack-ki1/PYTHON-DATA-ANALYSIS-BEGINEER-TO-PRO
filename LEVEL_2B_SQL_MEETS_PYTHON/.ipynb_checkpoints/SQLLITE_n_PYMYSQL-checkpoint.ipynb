## **1. SQLite: The Embedded Database**

### **What is SQLite?**
SQLite is a **serverless, zero-configuration, transactional SQL database engine** that:
- Stores entire database in a **single file** (or memory)
- Requires **no installation or server process**
- Supports **ACID transactions** (Atomicity, Consistency, Isolation, Durability)
- Is **public domain** (free for commercial use)
- Is the **most deployed database** in the world (used in Android, iOS, browsers, etc.)

> ðŸ’¡ **Key Insight**: SQLite is not a client-server databaseâ€”itâ€™s a **library** that reads/writes directly to disk.

### **When to Use SQLite**
âœ… **Ideal for**:
- Mobile apps and embedded systems
- Small-to-medium web applications (low write concurrency)
- Prototyping and development
- Local data caching
- Educational purposes

âŒ **Avoid for**:
- High-write concurrency (>100 writes/sec)
- Multi-user enterprise applications
- Very large datasets (>100GB)

---

### **Basic Setup & Connection**

#### **Import and Connect**

```python
import sqlite3

# Connect to file-based database (creates if doesn't exist)
conn = sqlite3.connect("ecommerce.db")

# Connect to in-memory database (for testing)
conn = sqlite3.connect(":memory:")
```

#### **Context Manager (Recommended)**
Automatically handles commit/rollback and closes connection:

```python
with sqlite3.connect("ecommerce.db") as conn:
    cur = conn.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS users (id INTEGER, name TEXT)")
    # Auto-commits on success, auto-rolls back on exception
```

#### **Connection Parameters**
| Parameter | Description |
|-----------|-------------|
| `database` | Path to DB file or `:memory:` |
| `timeout` | Seconds to wait for lock (default: 5.0) |
| `isolation_level` | `None` for autocommit, `"DEFERRED"` (default) for manual |
| `check_same_thread` | `False` to share connections across threads |

### **CRUD Operations**

#### **Create Table**

```python
with sqlite3.connect("app.db") as conn:
    cur = conn.cursor()
    
    # Enable foreign keys (disabled by default)
    cur.execute("PRAGMA foreign_keys = ON")
    
    cur.execute("""
        CREATE TABLE IF NOT EXISTS users (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            username TEXT UNIQUE NOT NULL,
            email TEXT NOT NULL,
            created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        )
    """)
    
    cur.execute("""
        CREATE TABLE IF NOT EXISTS posts (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_id INTEGER NOT NULL,
            title TEXT NOT NULL,
            content TEXT,
            FOREIGN KEY (user_id) REFERENCES users (id) ON DELETE CASCADE
        )
    """)
```

#### **Insert Data**

```python
# Single insert
cur.execute(
    "INSERT INTO users (username, email) VALUES (?, ?)",
    ("alice", "alice@example.com")
)

# Get last inserted ID
user_id = cur.lastrowid

# Multiple inserts
users_data = [
    ("bob", "bob@example.com"),
    ("charlie", "charlie@example.com")
]
cur.executemany(
    "INSERT INTO users (username, email) VALUES (?, ?)",
    users_data
)
```

#### **Select Data**

```python
# Fetch all
cur.execute("SELECT * FROM users")
all_users = cur.fetchall()

# Fetch one
cur.execute("SELECT * FROM users WHERE id = ?", (1,))
user = cur.fetchone()

# Fetch many
cur.execute("SELECT * FROM users")
batch = cur.fetchmany(10)

# Row factory for dict-like access
conn.row_factory = sqlite3.Row
cur = conn.cursor()
cur.execute("SELECT username, email FROM users")
row = cur.fetchone()
print(row["username"])  # Instead of row[0]
```

#### **Update & Delete**

```python
# Update
cur.execute(
    "UPDATE users SET email = ? WHERE username = ?",
    ("newemail@example.com", "alice")
)

# Delete
cur.execute("DELETE FROM users WHERE id = ?", (1,))
```

### **Advanced Features**

#### **Transactions**

```python
try:
    conn.execute("BEGIN IMMEDIATE")  # Acquire write lock immediately
    
    # Multiple operations
    conn.execute("UPDATE accounts SET balance = balance - 100 WHERE id = 1")
    conn.execute("UPDATE accounts SET balance = balance + 100 WHERE id = 2")
    
    conn.commit()
except Exception as e:
    conn.rollback()
    raise e
```

#### **Custom Functions**

```python
import hashlib

def md5(text):
    return hashlib.md5(text.encode()).hexdigest()

conn.create_function("md5", 1, md5)
cur.execute("SELECT md5(username) FROM users")
```

#### **Aggregates**

```python
class AvgLength:
    def __init__(self):
        self.total_len = 0
        self.count = 0
    
    def step(self, value):
        if value:
            self.total_len += len(value)
            self.count += 1
    
    def finalize(self):
        return self.total_len / self.count if self.count else 0

conn.create_aggregate("avg_length", 1, AvgLength)
cur.execute("SELECT avg_length(username) FROM users")
```

#### **Backup Database**

```python
def progress(status, remaining, total):
    print(f"Backed up {total-remaining} of {total} pages...")

with sqlite3.connect("live.db") as src:
    with sqlite3.connect("backup.db") as dst:
        src.backup(dst, pages=1, progress=progress)
```

---

### **Best Practices**

1. **Use parameterized queries** to prevent SQL injection
2. **Enable foreign keys**: `PRAGMA foreign_keys = ON`
3. **Use context managers** (`with` statements) for automatic cleanup
4. **Set row factory** for readable code: `conn.row_factory = sqlite3.Row`
5. **Handle locks gracefully** with `timeout` parameter
6. **Vacuum periodically** to reduce file size: `conn.execute("VACUUM")`

## **2. PyMySQL: MySQL Connector for Python**

### **What is PyMySQL?**
PyMySQL is a **pure Python MySQL client library** that:
- Implements the **Python DB-API 2.0** specification
- Is **compatible with MySQLdb** (drop-in replacement)
- Supports **Python 3** (unlike older MySQLdb)
- Works with **MySQL, MariaDB, and Percona Server**
- Is **MIT licensed** (free for commercial use)

> ðŸ’¡ **Key Insight**: PyMySQL communicates directly with MySQL servers using the MySQL wire protocolâ€”no C dependencies required.

### **When to Use PyMySQL**
âœ… **Ideal for**:
- Web applications (Django, Flask)
- Data pipelines connecting to MySQL
- Environments where you canâ€™t install C libraries (e.g., AWS Lambda)
- Applications requiring pure Python dependencies

âŒ **Consider alternatives**:
- `mysql-connector-python` (official Oracle driver) for advanced features
- `aiomysql` for async applications

### **Connection Management**

#### **Basic Connection**

```python
import pymysql

conn = pymysql.connect(
    host="localhost",
    user="myuser",
    password="mypassword",
    database="mydb",
    charset="utf8mb4",
    cursorclass=pymysql.cursors.DictCursor  # Returns dicts instead of tuples
)
```

#### **Connection Parameters**
| Parameter | Description |
|-----------|-------------|
| `host` | MySQL server address (default: "localhost") |
| `port` | Port number (default: 3306) |
| `user` | Username |
| `password` | Password |
| `database` | Database name |
| `charset` | Character set (use `"utf8mb4"` for full Unicode) |
| `cursorclass` | `pymysql.cursors.DictCursor` for dict results |
| `autocommit` | `True` to auto-commit (default: `False`) |
| `ssl` | SSL configuration dict |

#### **Context Manager (Recommended)**

```python
from contextlib import closing

with closing(pymysql.connect(...)) as conn:
    with conn.cursor() as cur:
        cur.execute("SELECT * FROM users")
        result = cur.fetchall()
    conn.commit()  # Explicit commit
```

#### **Connection Pooling (For Web Apps)**

```python
from pymysqlpool import ConnectionPool

config = {
    "host": "localhost",
    "user": "myuser",
    "password": "mypassword",
    "database": "mydb",
    "charset": "utf8mb4"
}

pool = ConnectionPool(size=5, name="mypool", **config)

with pool.connection() as conn:
    with conn.cursor() as cur:
        cur.execute("SELECT VERSION()")
        print(cur.fetchone())
```

### **CRUD Operations**

#### **Create Table**

```python
with pymysql.connect(...) as conn:
    with conn.cursor() as cur:
        cur.execute("""
            CREATE TABLE IF NOT EXISTS users (
                id INT AUTO_INCREMENT PRIMARY KEY,
                username VARCHAR(50) UNIQUE NOT NULL,
                email VARCHAR(100) NOT NULL,
                created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
            ) ENGINE=InnoDB
        """)
        conn.commit()  # Required for DDL in autocommit=False mode
```

#### **Insert Data**

```python
# Single insert
with conn.cursor() as cur:
    sql = "INSERT INTO users (username, email) VALUES (%s, %s)"
    cur.execute(sql, ("alice", "alice@example.com"))
    user_id = conn.insert_id()  # Get auto-increment ID
conn.commit()

# Multiple inserts (executemany uses efficient bulk insert)
users_data = [("bob", "bob@example.com"), ("charlie", "charlie@example.com")]
with conn.cursor() as cur:
    cur.executemany(sql, users_data)
conn.commit()
```

#### **Select Data**

```python
# With DictCursor
with conn.cursor(pymysql.cursors.DictCursor) as cur:
    cur.execute("SELECT username, email FROM users WHERE id = %s", (1,))
    user = cur.fetchone()
    print(user["username"])  # Readable!

# Pagination
with conn.cursor() as cur:
    offset = 0
    limit = 10
    while True:
        cur.execute("SELECT * FROM users LIMIT %s OFFSET %s", (limit, offset))
        rows = cur.fetchall()
        if not rows:
            break
        process_rows(rows)
        offset += limit
```

#### **Update & Delete**

```python
# Update with affected rows check
with conn.cursor() as cur:
    cur.execute(
        "UPDATE users SET email = %s WHERE username = %s",
        ("newemail@example.com", "alice")
    )
    if cur.rowcount == 0:
        print("No user found!")
conn.commit()

# Safe delete with confirmation
with conn.cursor() as cur:
    cur.execute("SELECT COUNT(*) FROM orders WHERE user_id = %s", (user_id,))
    if cur.fetchone()[0] > 0:
        raise ValueError("Cannot delete user with orders!")
    cur.execute("DELETE FROM users WHERE id = %s", (user_id,))
conn.commit()
```

### **Advanced Features**

#### **Transactions**

```python
try:
    conn.begin()  # Start transaction
    
    with conn.cursor() as cur:
        cur.execute("UPDATE accounts SET balance = balance - 100 WHERE id = 1")
        cur.execute("UPDATE accounts SET balance = balance + 100 WHERE id = 2")
    
    conn.commit()
except Exception as e:
    conn.rollback()
    raise e
```

#### **Stored Procedures**

```python
# Create procedure
with conn.cursor() as cur:
    cur.execute("""
        CREATE PROCEDURE GetUserInfo(IN user_id INT)
        BEGIN
            SELECT username, email FROM users WHERE id = user_id;
        END
    """)
    conn.commit()

# Call procedure
with conn.cursor() as cur:
    cur.callproc("GetUserInfo", (1,))
    result = cur.fetchone()
```

#### **SSL Connections**

```python
ssl_config = {
    "ca": "/path/to/ca.pem",
    "cert": "/path/to/client-cert.pem",
    "key": "/path/to/client-key.pem"
}

conn = pymysql.connect(
    host="secure-db.example.com",
    ssl=ssl_config,
    ...
)
```

#### **Read Replicas**

```python
# Primary for writes
primary_conn = pymysql.connect(host="primary-db", ...)

# Replica for reads
replica_conn = pymysql.connect(host="replica-db", ...)

# Write to primary
with primary_conn.cursor() as cur:
    cur.execute("INSERT INTO logs (message) VALUES (%s)", ("New log",))
primary_conn.commit()

# Read from replica
with replica_conn.cursor() as cur:
    cur.execute("SELECT COUNT(*) FROM logs")
    count = cur.fetchone()[0]
```

### **Best Practices**

1. **Always use parameterized queries** (`%s` placeholders)
2. **Set `charset="utf8mb4"`** for full Unicode support (emojis!)
3. **Use `DictCursor`** for readable code
4. **Explicitly commit transactions** (autocommit=False by default)
5. **Handle connection errors**:

```python
    from pymysql.err import OperationalError
   
   try:
       cur.execute("SELECT * FROM users")
   except OperationalError as e:
       if e.args[0] == 2006:  # MySQL server gone away
           reconnect()
   
```

6. **Use connection pooling** for web applications

## **3. Comparison: SQLite vs PyMySQL**

| Feature | SQLite | PyMySQL |
|---------|--------|---------|
| **Architecture** | Embedded (file-based) | Client-server |
| **Concurrency** | Limited write concurrency | High concurrency |
| **Setup** | Zero configuration | Requires MySQL server |
| **Scalability** | Up to ~100GB | Terabytes+ |
| **Network** | Local only | Network accessible |
| **Data Types** | Dynamic typing | Strict typing |
| **Use Case** | Mobile, prototyping, small apps | Web apps, enterprise systems |
| **Installation** | Built into Python | `pip install PyMySQL` |
| **Transactions** | Full ACID | Full ACID |
| **Security** | File permissions | User accounts, SSL |

## **4. Real-World Examples**

### **Example 1: SQLite for Local Caching**

```python
import sqlite3
import requests
from datetime import datetime, timedelta

class APICache:
    def __init__(self, db_path="cache.db"):
        self.conn = sqlite3.connect(db_path)
        self._create_table()
    
    def _create_table(self):
        self.conn.execute("""
            CREATE TABLE IF NOT EXISTS api_cache (
                url TEXT PRIMARY KEY,
                response TEXT,
                timestamp TIMESTAMP
            )
        """)
    
    def get(self, url, ttl_minutes=60):
        cutoff = datetime.now() - timedelta(minutes=ttl_minutes)
        cur = self.conn.execute(
            "SELECT response FROM api_cache WHERE url = ? AND timestamp > ?",
            (url, cutoff)
        )
        row = cur.fetchone()
        return row[0] if row else None
    
    def set(self, url, response):
        self.conn.execute(
            "INSERT OR REPLACE INTO api_cache (url, response, timestamp) VALUES (?, ?, ?)",
            (url, response, datetime.now())
        )
        self.conn.commit()

# Usage
cache = APICache()
data = cache.get("https://api.example.com/data")
if not data:
    data = requests.get("https://api.example.com/data").text
    cache.set("https://api.example.com/data", data)
```

### **Example 2: PyMySQL for Web Application**

```python
import pymysql
from flask import Flask, jsonify, request

app = Flask(__name__)

def get_db_connection():
    return pymysql.connect(
        host="localhost",
        user="webapp",
        password="secure_password",
        database="webapp_db",
        charset="utf8mb4",
        cursorclass=pymysql.cursors.DictCursor,
        autocommit=True
    )

@app.route("/users/<int:user_id>")
def get_user(user_id):
    with get_db_connection() as conn:
        with conn.cursor() as cur:
            cur.execute("SELECT id, username, email FROM users WHERE id = %s", (user_id,))
            user = cur.fetchone()
            if user:
                return jsonify(user)
            else:
                return jsonify({"error": "User not found"}), 404

@app.route("/users", methods=["POST"])
def create_user():
    data = request.json
    with get_db_connection() as conn:
        with conn.cursor() as cur:
            try:
                cur.execute(
                    "INSERT INTO users (username, email) VALUES (%s, %s)",
                    (data["username"], data["email"])
                )
                return jsonify({"id": conn.insert_id(), "username": data["username"]}), 201
            except pymysql.IntegrityError as e:
                if "Duplicate entry" in str(e):
                    return jsonify({"error": "Username already exists"}), 409
                raise e
```
