```python
#HOW TO CREATE A NEW  COLUMN USING A FUNCTION 
def Age_Group(Age):
    if Age <=20:
        return 'Minor'
    elif  21 <= Age < 40:
        return 'Adult'
    else:
        return 'Old'
df["Age_Group"]=df["Age"].apply(Age_Group)
df
```
