### **Data Structure Creation**
*   **`pd.Series()`**: Creates a one-dimensional labeled array capable of holding any data type.
*   **`pd.DataFrame()`**: Creates a two-dimensional, size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns).
*   **`pd.Index()`**: Creates a pandas Index object for holding axis labels.
*   **`pd.MultiIndex()`**: Used for creating hierarchical (multi-level) indexes.

### **Data Inspection and Exploration**
*   **`.head(n)`** / **`.tail(n)`**: Returns the first or last $n$ rows of a object.
*   **`.info()`**: Provides a concise summary of a DataFrame, including index, dtypes, and non-null values.
*   **`.describe()`**: Generates descriptive statistics that summarize the central tendency, dispersion, and shape of a datasetâ€™s distribution.
*   **`.shape`**: Attribute that returns a tuple representing the dimensionality of the DataFrame (rows, columns).
*   **`.dtypes`**: Returns the data types of each column.
*   **`.index`** / **`.columns`**: Accessors for the row labels and column names.
*   **`.values`** or **`.to_numpy()`**: Returns a NumPy representation of the data.
*   **`.size`**: Returns the number of elements in an object.
*   **`.count()`**: Returns the number of non-missing values in each column or row.
*   **`.value_counts()`**: Returns a Series containing counts of unique values.
*   **`.nunique()`** / **`.unique()`**: Counts or lists unique values in a Series.

### **Data Selection and Subsetting**
*   **`[]` (Square Brackets)**: Primarily used for selecting columns by name or slicing rows.
*   **`.loc[]`**: Accesses a group of rows and columns by **labels** or a boolean array.
*   **`.iloc[]`**: Accesses a group of rows and columns by **integer positions**.
*   **`.at[]`** / **`.iat[]`**: Fast accessors for getting or setting a **single scalar value** by label or position.
*   **`.query()`**: Filters a DataFrame using a boolean expression string.
*   **`.get_group()`**: Used with GroupBy objects to retrieve a specific group as a DataFrame.
*   **`.filter()`**: Filters rows or columns of a DataFrame based on their labels.

### **Data Cleaning and Missing Values**
*   **`pd.isnull()`** or **`.isna()`**: Detects missing values (NaN or None).
*   **`pd.notnull()`** or **`.notna()`**: Detects non-missing values.
*   **`.fillna()`**: Fills NA/NaN values using the specified method (e.g., constant value, forward fill, or backward fill).
*   **`.dropna()`**: Removes missing values from an axis.
*   **`.interpolate()`**: Fills missing values using various **interpolation** methods (like linear).
*   **`.duplicated()`**: Identifies duplicate rows.
*   **`.drop_duplicates()`**: Removes duplicate rows.
*   **`.replace()`**: Replaces specific values with other values.

### **Transformation and Manipulation**
*   **`.drop()`**: Removes specified labels from rows or columns.
*   **`.assign()`**: Assigns new columns to a DataFrame, supporting method chaining.
*   **`.astype()`**: Casts a pandas object to a specified **dtype**.
*   **`.rename()`**: Alters axes labels (indexes or columns).
*   **`.apply()`**: Applies a function along an axis of the DataFrame or Series.
*   **`.map()`**: Map values of a Series according to input correspondence (dictionary or function).
*   **`.applymap()`**: Applies a function to a Dataframe **element-wise**.
*   **`.copy()`**: Creates a deep copy of the object.
*   **`.pop()`**: Removes an item and returns it (deletes a column).
*   **`.set_index()`** / **`.reset_index()`**: Sets the DataFrame index using existing columns or resets the index to default.

### **Merging, Joining, and Concatenating**
*   **`pd.merge()`**: Merges DataFrame objects using a **database-style join**.
*   **`pd.concat()`**: Concatenates pandas objects along a particular axis.
*   **`.join()`**: Joins columns with another DataFrame either on index or on a key column.
*   **`.combine_first()`**: Combines two objects by filling nulls in the first with values from the second.

### **Reshaping and Pivoting**
*   **`.pivot()`**: Reshapes data from "long" to "wide" format based on column values.
*   **`.pivot_table()`**: Creates a spreadsheet-style pivot table as a DataFrame.
*   **`pd.melt()`**: Unpivots a DataFrame from wide to long format.
*   **`.stack()`** / **`.unstack()`**: Pivots a level of the column labels to the index (stack) or vice versa (unstack).
*   **`pd.crosstab()`**: Computes a simple cross-tabulation of two (or more) factors.
*   **`pd.get_dummies()`**: Converts categorical variables into **dummy/indicator variables** (one-hot encoding).
*   **`pd.cut()`** / **`pd.qcut()`**: Bins continuous data into discrete intervals based on values or quantiles.

### **Aggregation and Grouping**
*   **`.groupby()`**: Groups data using a mapper or by a Series of columns for "split-apply-combine" operations.
*   **`.agg()`** (or **`.aggregate()`**): Aggregates using one or more operations over the specified axis.
*   **`.transform()`**: Calls a function on self producing a DataFrame with transformed values (same shape as input).

### **Statistical Methods**
*   **`.sum()`**, **`.mean()`**, **`.median()`**, **`.mode()`**: Basic central tendency calculations.
*   **`.std()`** / **`.var()`**: Computes standard deviation and variance.
*   **`.min()`** / **`.max()`**: Identifies minimum and maximum values.
*   **`.quantile()`**: Returns values at the given quantile over a requested axis.
*   **`.cumsum()`** / **`.cumprod()`**: Computes cumulative sum or product.
*   **`.pct_change()`**: Computes the fractional change between the current and a prior element.
*   **`.diff()`**: First discrete difference of element.
*   **`.corr()`** / **`.cov()`**: Computes pairwise correlation or covariance of columns.
*   **`.corrwith()`**: Computes pairwise correlation between rows or columns of two DataFrame objects.
*   **`.rank()`**: Computes numerical data ranks (1 through n).
*   **`.idxmin()`** / **`.idxmax()`**: Returns the index of the first occurrence of the minimum or maximum.

### **Time Series Handling**
*   **`pd.to_datetime()`**: Converts argument to **datetime**.
*   **`pd.date_range()`**: Returns a fixed frequency **DatetimeIndex**.
*   **`pd.period_range()`**: Returns a fixed frequency **PeriodIndex**.
*   **`pd.Period()`**: Represents a period of time.
*   **`.resample()`**: Convenience method for frequency conversion and resampling of time series.
*   **`.shift()`**: Shifts index by desired number of periods with an optional time freq.
*   **`.tz_localize()`** / **`.tz_convert()`**: Localizes a time zone-naive index or converts a time zone-aware index.
*   **`.to_period()`** / **`.to_timestamp()`**: Converts between timestamp-based and period-based indexes.
*   **`.asof()`**: Returns the last labels at or before a specific date.
*   **`.rolling()`**: Provides rolling window calculations.
*   **`.expanding()`**: Provides expanding transformations.
*   **`.ewm()`**: Provides exponential weighted functions.
*   **`.at_time()`** / **`.between_time()`**: Selects values at a particular time or between specific times.

### **File I/O (Input/Output)**
*   **`pd.read_csv()`** / **`.to_csv()`**: Reads/Writes a comma-separated values (CSV) file.
*   **`pd.read_table()`**: Reads a general delimited file into a DataFrame.
*   **`pd.read_excel()`** / **`.to_excel()`**: Reads/Writes an Excel file.
*   **`pd.read_json()`** / **`.to_json()`**: Reads/Writes a JSON string/file.
*   **`pd.read_html()`** / **`.to_html()`**: Reads/Writes HTML tables.
*   **`pd.read_pickle()`** / **`.to_pickle()`**: Reads/Writes a pickled object to disk.
*   **`pd.read_sql()`** / **`.to_sql()`**: Reads/Writes from a SQL database.
*   **`pd.read_xml()`** / **`.to_xml()`**: Reads/Writes XML data.
*   **`pd.read_feather()`** / **`.to_feather()`**: Reads/Writes the binary Feather format.
*   **`pd.read_parquet()`** / **`.to_parquet()`**: Reads/Writes the Parquet binary format.

### **Sorting**
*   **`.sort_index()`**: Sorts object by labels along an axis.
*   **`.sort_values()`**: Sorts by the values along either axis.
*   **`.nsmallest()`** / **`.nlargest()`**: Returns the $n$ smallest or largest elements.

### **Experimental/Special Tools**
*   **`pd.NA`**: Represents a experimental, missing value that works across various types.
*   **`pd.Grouper()`**: Allows for complex grouping specifications, especially for time frequencies.
